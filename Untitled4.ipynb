# Problem Statement

AllLife Bank is a US bank that has a growing customer base. The majority of these customers are liability customers (depositors) with varying sizes of deposits. The number of customers who are also borrowers (asset customers) is quite small, and the bank is interested in expanding this base rapidly to bring in more loan business and in the process, earn more through the interest on loans. In particular, the management wants to explore ways of converting its liability customers to personal loan customers (while retaining them as depositors).

A campaign that the bank ran last year for liability customers showed a healthy conversion rate of over 9% success. This has encouraged the retail marketing department to devise campaigns with better target marketing to increase the success ratio.

You as a Data scientist at AllLife bank have to build a model that will help the marketing department to identify the potential customers who have a higher probability of purchasing the loan.

### Objective

- To predict whether a liability customer will buy a personal loan or not.
- Which variables are most significant.
- Which segment of customers should be targeted more.

### Data

* ID: Customer ID
* Age: Customer’s age in completed years
* Experience: #years of professional experience
* Income: Annual income of the customer (in thousand dollars)
* ZIP Code: Home Address ZIP code.
* Family: the Family size of the customer
* CCAvg: Average spending on credit cards per month (in thousand dollars)
* Education: Education Level. 1: Undergrad; 2: Graduate;3: Advanced/Professional
* Mortgage: Value of house mortgage if any. (in thousand dollars)
* Personal_Loan: Did this customer accept the personal loan offered in the last campaign?
* Securities_Account: Does the customer have securities account with the bank?
* CD_Account: Does the customer have a certificate of deposit (CD) account with the bank?
* Online: Do customers use internet banking facilities?
* CreditCard: Does the customer use a credit card issued by any other Bank (excluding All life Bank)?

Personal_Loan is the dependent variable. We have 13 independent variables and 1 dependent variables we want to model. Since ID is likely not useful, we have 12 independent variables.

```python
X = data.drop(columns=['Personal_Loan'])
y = data['Personal_Loan']
```

```python
y_prob = logistic.predict_proba(X_test)[:, 1]
[fpr, tpr, thresholds] = roc_curve(y_test, y_prob)
pr_diff = tpr-fpr
opt_thresh_index = np.argmax(pr_diff)
plt.plot(fpr, tpr)
plt.scatter(fpr[opt_thresh_index], tpr[opt_thresh_index])
opt_thresh = thresholds[opt_thresh_index]
print("best threshold is: ", opt_thresh)
```

```python
optimize_recall = opt_thresh
preds = y_prob > opt_thresh
print(recall_score(y_test, preds))
print(precision_score(y_test, preds))
print(roc_auc_score(y_test, preds))
```

```python
logistic = LogisticRegression(solver="liblinear")
logistic.fit(X_train, y_train)
probs = logistic.predict_proba(X_test)[:, 1]
[precision, recall, thresholds] = precision_recall_curve(y_test, probs)
precision += 0.000001
recall += 0.00001
f1_scores = (2 * precision * recall ) / (precision + recall)
best_threshold_index = np.argmax(f1_scores)
plt.plot(recall[:-1], precision[:-1], marker='.', label="logistic")
plt.scatter(recall[best_threshold_index], precision[best_threshold_index], c="red")
plt.show()
print(thresholds[best_threshold_index])
plt.plot(thresholds, precision[:-1], "b--", label="precision")
plt.plot(thresholds, recall[:-1], "g--", label="recall")
plt.legend(loc="upper left")
plt.show()
```

```python
precision_recall_balance = thresholds[best_threshold_index]
preds = y_prob > thresholds[best_threshold_index]
print(recall_score(y_test, preds))
print(precision_score(y_test, preds))
```

```python
def compare_thresholds(*thresholds, labels=None):
    results = []
    for t in thresholds:
        preds = y_prob > t
        d = pd.DataFrame([[recall_score(y_test, preds), precision_score(y_test, preds), f1_score(y_test, preds)]], columns=["recall", "precision", "f1"])
        results.append(d)
    results = pd.concat(results)
    results.index=labels
    return results
```

```python
compare_thresholds(optimize_recall, precision_recall_balance, labels=["roc", "precision_recall"])
```

The ROC threshold of 12.5% has the best threshold for recall, which is suitable for this context. We want to find the most people who are likely to get a loan, and advertise towards them.

### Interpretation

The coefficients are linearly proportional to the log odds of the independent variables. Therefore, we see that Income, Education, and Family have the greatest effect on the odds of a liability customer getting a loan.
Every unit increase in the sqrt of the income increases the odds by 77.3 percent.
Every unit increase in the years of education increases the odds by 19.8 percent.
Every unit increase in the number of family members increases the odds by 14.7 percent.

Interestingly, having a mortgage decreases the odds of getting a loan slightly.
Interestingly, age and experience have little effect on the odds of getting a loan.

# Decision Tree

```python
def plot_pretty_tree(dTree, figsize=(20,30)):
    plt.figure(figsize=figsize)
    out = plot_tree(
        dTree,
        feature_names=X_train.columns,
        filled=True,
        fontsize=9,
        node_ids=True,
        class_names=True,
    )
    for o in out:
        arrow = o.arrow_patch
        if arrow is not None:
            arrow.set_edgecolor("black")
            arrow.set_linewidth(1)
    plt.show()
    
def make_confusion(dTree):
    cm = confusion_matrix(y_test, dTree.predict(X_test), labels=[0,1])
    labels = np.asarray(
            [
                ["%0.0f \n %0.2f%s" % (item, 100 * item / (cm.flatten().sum()), "%")]
                for item in cm.flatten()
            ]
        ).reshape(2, 2)
    sns.heatmap(cm, annot=np.array(labels), fmt = '')
    
def show_scores(dTree):
    preds = dTree.predict(X_test)
    f1 = f1_score(y_test, preds)
    recall = recall_score(y_test, preds)
    accuracy = dTree.score(X_test, y_test)
    precision = precision_score(y_test, preds)
    return pd.DataFrame([[accuracy, recall, precision, f1]], columns=["acc", "recall", "precision", "f1"])

def compare_trees(*trees, labels=None):
    scores = []
    for t in trees:
        scores.append(show_scores(t))
    result = pd.concat(scores)
    if labels:
        result.index = labels
    return result
```

```python
def plot_importances(dTree):
    importances = pd.concat([pd.DataFrame(X_train.columns), pd.DataFrame(dTree.feature_importances_)], axis=1)
    importances.columns=["cols", "importances"]
    importances = importances.sort_values(by="importances")
    sns.barplot(x = "importances", y="cols", data=importances)
```

## Business Insights

The decision tree gets much better scores for accuracy, recall, and precision. However, logistic regression is still useful because of its interpretability factor. Also, it is likely more generalizable if new data causes the decision tree to drastically change structure. From both models, the following conclusions can be reached.

The three most important factors in someone requesting a loan is their Education, Income, and Family. This was true across all three versions of the decision tree, as well as logistic regression.

Having more education seems to lead to "no" in loans, same with having more family.
Someone who makes more than 115,000 and has more than 1.5 years of school is very likely to not get a loan.
Meanwhile, someone who makes more than 115,000 and has less than 1.5 years of school and less than 2.5 family members is likely to get a loan. Someone who makes the same amount, same schooling, but has more than 2.5 family members is likely to get a loan.

Someone who has an income of less than 100,000 is likely to get a loan.

Customer segments that are worth considering are people with more than 2 family members, income less than 115,000, and less than 1.5 years of school.
